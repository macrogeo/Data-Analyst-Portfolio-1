```python
# Web scraping project
# Collect data from shopee website
# As a seller, we have to know the market
# - What keywords other seller use
# - How other seller post photos
# - How other boost their sales
```

```python
# Get started with Selenium by downloading ChromeDriver from https://chromedriver.chromium.org/downloads
# Install Selenium and implement webdriver.Chrome() to open browser.
#!pip install selenium
from selenium import webdriver
driver = webdriver.Chrome(executable_path=r'C:\Users\GL\Downloads\WebScrapingProject-Shopee\chromedriver.exe')
```

```python
# Get data from the web destination
driver.get('https://shopee.co.th/')
```

```python
# By using selenium means we will use commands to gather data from the target website without clicking anything. It's like  we are trying to mimic our behavior when we use the website using only the commands line.
# Therefore, we have to inspect the html code to get the address of what we are looking for
# We will select Thai language - copy as Full Xpath
thai_button = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div/div[3]/div[1]/button')
```

```python
# Order python to click on Thai language
thai_button.click()
```

```python
# After selecting language Shopee will pop up some ad, normally, we will click the close button to close it but for data scraping, we have to order python to do it.
# The point is, from inspection, ad close button is under shadow root which makes python unable to access it.
# So we will access the div that contain shadow root and write the command to get the data in that div
# We will use python command with Javascript
close_button = driver.execute_script('return document.querySelector("shopee-banner-popup-stateful").shadowRoot.querySelector("div.shopee-popup__close-btn")')
```

```python
close_button.click()
```

```python
# Go to search box
search = driver.find_elements_by_xpath(r'/html/body/div[1]/div/header/div[2]/div/div[1]/div[1]/div/form/input')
```

```python
search[0].send_keys('โต๊ะปรับระดับไฟฟ้า')
```

```python
# To make python click enter after search we will import more libraries
from selenium.webdriver.common.keys import Keys
```

```python
search[0].send_keys(Keys.ENTER)
```

```python
# After the result shows up. It can be seen that some results didn't show up if we do not scroll down
# To solve this issue we will use python with Javascript to zoom out
driver.execute_script("document.body.style.zoom='10%'")
```

```python
# Let's getting data and save it as dataframe
data = driver.page_source
```

```python
#Using BeautifulSoup
import bs4
soup = bs4.BeautifulSoup(data)
soup
```

```python
# Now we have the dataset in a local computer so we can zoom out
driver.execute_script("document.body.style.zoom='100%'")
```

```python
# Select the data we want to gather
# - Product Name using class = ie3A+n bM+7UW Cve6sh
all_prodect_1st_page = soup.find_all('div',{'class':'ie3A+n bM+7UW Cve6sh'})
```

```python
# Check how many product in the first page
len(all_prodect_1st_page)
# So Shopee have 60 products for the first page
```

```python
# Get the name of all products
all_prodect_1st_page_list = []
for product in all_prodect_1st_page:
    all_prodect_1st_page_list.append(product.text)
all_prodect_1st_page_list
```

```python
# Get the price
all_price_1st_page = soup.find_all('div',{'class':'vioxXd rVLWG6'})
all_price_1st_page
```

```python
# Get the price of all products
all_price_1st_page_list = []
for price in all_price_1st_page:
    all_price_1st_page_list.append(price.text)
all_price_1st_page_list
```

```python
# How many sales r6HknA uEPGHT
all_sales_1st_page = soup.find_all('div',{'class':'r6HknA uEPGHT'})
all_sales_1st_page
```

```python
# Get the sales of all products
all_sales_1st_page_list = []
for sales in all_sales_1st_page:
    all_sales_1st_page_list.append(sales.text)
all_sales_1st_page_list
```

```python
# Use pandas to get the output as excel files
import pandas as pd
```

```python
shopee_data = pd.DataFrame([all_prodect_1st_page_list,all_price_1st_page_list,all_sales_1st_page_list])
shopee_data
```

```python
# Transpose data to make it easier to read
shopee_data_transpose = shopee_data.transpose()
shopee_data_transpose
```

```python
# Change index name
shopee_data_transpose.columns = ['Title','Price','Total Sales']
shopee_data_transpose
```

```python
# Export data to excel
shopee_data_transpose.to_excel(r'C:\Users\GL\Downloads\WebScrapingProject-Shopee\Shopee_โต๊ะปรับระดับ.xlsx')
```
