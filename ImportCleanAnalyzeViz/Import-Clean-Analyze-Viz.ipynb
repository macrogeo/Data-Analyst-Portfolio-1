```python
# Data preparation
# Actually, it have many way to clean the data for example
# 1.drop all the missing values
```

```python
#Example 1
    # we will use taxi dataset as a 1st example
    # we know that taxi dataset have 6433 rows × 14 columns
    # then we have to check for missing values. Today we will use pandas tools
is_NaN = taxi.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = taxi[row_has_NaN]
rows_with_NaN
```

```python
# From above taxi have 92 rows that contain missing value
# Now we going to drop all missing values
taxi_no_missing = taxi.dropna()
taxi_no_missing
# The result show that now we have 6341 rows without missing values
```

```python
# 2. In case we didn't want to delete data that contains missing values
#    We can replace NaN with some value else by using fillna() instead of dropna()
    #To do so we will 
#    1. Split data we want to fill because we might struggle with a different data types in each column that can't fill with numeric values or strings
#    2. Manipulate the data
#    3. Combine all manipulated data together to create new dataframe
```

```python
#Example 2
# we will use crypto dataset
crypto
```

```python
# From above we can see that every row contains NaN therefore if we use
# crypto.dropna() it will remove all the data
# Therefore, we can set the condition to prevent it by using tresh, tresh will allows you to set the drop condition
crypto.dropna(thresh=3) 
# thresh = 3 means if the data rows contain NaN more than 3 NaN, the rows will be drop
```

```python
# To delete the column we will use dropna(axis=1)
#when we use dropna() the default value in parentheses will be axis=0 which will drop rows
#But when we set dropna(axis = 1) it will delete column
crypto.dropna(thresh=3).dropna(axis=1)
```

```python
# After finished cleaning the data now we're going to analyze it
# Using taxi_no_missing dataset as example
taxi_no_missing
```

```python
# Example3
#1. We will spilt the data using .loc[] to select row and column
#for example .loc[:,:] means select all rows and all columns
taxi.loc[[0,4],['fare','tip','tolls','total']] #will choose row 0,4 and column 'fare','tip','tolls','total'
```

```python
taxi.loc[0:9,['fare','tip','tolls','total']]
#will choose row 0 to 9 and column 'fare','tip','tolls','total'
```

```python
# We want to convert currency from usd to thb
# First we will create new table that contain all price in usd
price_usd = taxi.loc[:,['fare','tip','tolls','total']]
price_usd
```

```python
#Convert currency from usd to thb 
#1 United States Dollar equals 35.49 Thai Baht @Jun 22, 06:53 UTC
price_thb = price_usd * 35.49
price_thb
```

```python
# Next find price per distance (usd/miles)
price_distance = taxi['total']/taxi['distance']
price_distance
```

```python
# After finish transformed data we will put it back to taxi_no_missing
    #Method 1: add column in table
taxi_no_missing['price per distance'] = price_distance
taxi_no_missing
```

```python
#Method 2 merge table from price_thb to taxi_no_missing
#Note: best practice not to make the duplicate column name
#Change the column name in table price_thb
price_thb.columns
```

```python
price_thb.columns = ['fare_THB', 'tip_THB', 'tolls_THB', 'total_THB']
price_thb
```

```python
#Merge the table price_thb to table taxi_no_missing
taxi_no_missing.merge(price_thb, left_index=True, right_index=True)
```

```python
# Let's save the data as new dataframe
report = taxi.merge(price_thb, left_index=True, right_index=True)
report
```

```python
# Export the data as excel or csv using .to_excel or .to_csv
report.to_excel(r'C:\Users\GL\Downloads\UltimatePythonXExcel\ExcelFiles\Report.xlsx')
```

```python
#Find percentage of tips from total bill
tips_with_pct = tips
tips_with_pct['pct_tips'] = tips['tip']/tips['total_bill']
tips_with_pct
```

```python
sns.catplot(x='day', y='pct_tips',data = tips_with_pct,kind='box')
```

```python
#From box plot we can see that it have outlier that makes visualization unclear, so we will drop the outlier
tips_with_pct = tips_with_pct.sort_values(by='pct_tips',ascending=False)
tips_with_pct
```

```python
# Drop the outlier
tips_with_pct = tips_with_pct.sort_values(by='pct_tips',ascending=False).iloc[2:]
tips_with_pct
```

```python
sns.catplot(x='day', y='pct_tips',data = tips_with_pct,kind='box')
```

```python
#Let's do more filter
sns.catplot(x='day', y='pct_tips',data = tips_with_pct, kind='box',hue='sex')
```

```python
# Compare two boxplot between smoker
sns.catplot(x='day', y='pct_tips',data = tips_with_pct, kind='box',hue='sex',col='smoker')
```
